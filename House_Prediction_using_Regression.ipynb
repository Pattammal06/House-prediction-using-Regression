#SIMPLE LINEAR REGRESSION

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(32)
```

```python
area = np.random.randint(400,6000,100)
bedroom = np.random.randint(1,5,100)
washroom = np.random.randint(1,3,100)
age = np.random.randint(1,60,100)
```

```python
price = (area*0.06) + (bedroom*5) + (washroom*4) + (age*0.03) + np.random.normal(0,10,100)
```

```python
data = pd.DataFrame({
    "area" :area,
    "bedroom" :bedroom,
    "washroom" :washroom,
    "age" :age,
    "price" :price.round(2)
})
```

```python
data
```

```python
df=data
data
```

```python
x=df[["age"]]
y=df[["price"]]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("MAE : ",mean_absolute_error(y_test,y_pred))
print("MSE : ",mean_squared_error(y_test,y_pred))
print("RMSE : ",np.sqrt(mean_squared_error(y_test,y_pred)))
print("R2 SCORE : ",r2_score(y_test,y_pred))
```

```python
print("R2 SCORE : ",r2_score(y_test,y_pred))
```

```python
y_pred=model.predict(x_test)
```

```python
print("MAE:",mean_absolute_error(y_test,y_pred))
print("MSE:",mean_squared_error(y_test,y_pred))
print("RMSE:",root_mean_squared_error(y_test,y_pred))
print("R2 SCORE:",r2_score(y_test,y_pred))
```

```python
plt.figure(figsize=(6,5))
sns.kdeplot(y_test,label="ACTUAL PRICE",shade=True,color="green",alpha=0.5)
sns.kdeplot(y_pred,label="PREDICTED PRICE",fill=True,color="red",alpha=0.01)
plt.title("ACTUAL PRICE VS PREDICTED PRICE")
plt.xlabel("Score")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(range(len(y_test)),y_test.values)
plt.plot(range(len(y_pred)),y_pred)
plt.title("ACTUAL VS PREDICTED LINECHART")
plt.xlabel("INDEX")
plt.ylabel("PRICE")
plt.legend()
plt.grid()
plt.show()
```

```python
x=y_test.ravel()
y=y_pred.ravel()
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x=y_test,y=y_pred)
sns.kdeplot(y_test,label="ACTUAL PRICE",fill=True,color="green",alpha=0.8)
sns.kdeplot(y_pred,label="PREDICTED PRICE",fill=True,color="yellow",alpha=1)
plt.plot([y_test.min(),y_pred.max()],[y_test.min(),y_pred.max()],'r--')
plt.title("Regression Linechart")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
labels = ['MAE', 'MSE', 'RMSE', 'R2 Score']
plt.pie([mean_absolute_error(y_test,y_pred),mean_squared_error(y_test,y_pred),root_mean_squared_error(y_test,y_pred),r2_score(y_test,y_pred)], labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Regression Metrics")
plt.legend(loc="best")
plt.show()
```
